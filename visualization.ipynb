```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import seaborn as sns
```

```python
d = {'Algorithm': ['k-means','k-means','SVM','SVM','DBSCAN','DBSCAN'], 'Accuracy %': [63.20,58.79, 99,97.47, 72.833,70.70], 'Type':['train','test','train','test','train','test'  ]} 

df = pd.DataFrame(data=d)
df
```

```python
ax = sns.barplot(x="Algorithm", y="Accuracy %", hue="Type",color='#004488', data=df)
sns.set_style('whitegrid')
ax.set_ylim(45, 100)

figure = ax.get_figure()    
figure.savefig('/home/antara/Documents/researchproject/HDoutputs/accuracyresults.png', dpi=400)
```
